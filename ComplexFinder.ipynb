### Load data

```python
oct4_data_filepath = r'data\oct4_data.txt'
e7_data_filepath = r"data\e7_data.txt"
hgnc_data_filepath = r"data\complexes.txt"
```

```python
oct4_datatable = ComplexFinder.ppi_reader(data_filepath=oct4_data_filepath)
e7_datatable = ComplexFinder.ppi_reader(data_filepath=e7_data_filepath)
hgnc_dict = ComplexFinder.complex_reader(data_filepath=hgnc_data_filepath)
```

### Find Protein Complexes of Interest

```python
oct4_dict, e7_dict, common_dict = ComplexFinder.complex_finder(datatable_a=oct4_datatable, datatable_b=e7_datatable, complex_dictionary=hgnc_dict)
```

### Plot Protein Complexes of Interest

```python
ComplexFinder.complex_plotter(complex_finder_dictionary=oct4_dict, complex_reader_dictionary=hgnc_dict, title='Complexes with POU5F1 interactors.', font=8)
```

```python
ComplexFinder.complex_plotter(complex_finder_dictionary=e7_dict, complex_reader_dictionary=hgnc_dict, title='Complexes with E7 interactors.', font=10)
```

```python
ComplexFinder.complex_plotter(complex_finder_dictionary=common_dict, complex_reader_dictionary=hgnc_dict, title='Complexes with POU5F1 and E7 interactors.', font=12)
```
